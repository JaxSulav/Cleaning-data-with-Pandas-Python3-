```python
import pandas as pd
df = pd.read_csv('weather_data.csv', parse_dates=['day'])
df.set_index('day', inplace=True)
df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how="all") #drops row data which has NA in all columns
new_df
```

```python
new_df = df.dropna(thresh=1) #kept the row if at least 1 NA value. thresh =2 needs 2 NAs in that row for the row to be not dropped
new_df
```
