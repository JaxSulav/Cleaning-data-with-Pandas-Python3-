```python
df = pd.read_csv('weather_data.csv', parse_dates=['day'])
df.set_index('day', inplace=True) #inplace = true -it will modify the previous dataset, not doing this will return a new dataset
df
```

```python
new_df = df.fillna(0) #filling the nan value with zeros
new_df
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed': "zero",
    'event': "no_event"
})
new_df
```

```python
new_df = df.fillna(method="ffill") #fill the value same as previous value
new_df 
```

```python
new_df = df.fillna(method="bfill", axis = "columns") #fill the value horizontally
new_df 
```

```python
new_df = df.fillna(method="bfill", limit=1) #fill the previous value only once
new_df 
```
