```python
import pandas as pd
df = pd.read_csv('weather_data.csv', parse_dates=['day'])
df.set_index('day', inplace=True)
df
```

```python
new_df = df.interpolate() #linear interpolation
new_df
```

```python
new_df = df.interpolate(method='time') #linear interpolation as per time
new_df
```

```python
new_df = df.interpolate(method='quadratic') 
new_df
```

```python
new_df = df.interpolate(method='krogh') 
new_df
```

```python
new_df = df.interpolate(method='piecewise_polynomial')
new_df
```

```python
new_df = df.interpolate(method='from_derivatives')
new_df
```
